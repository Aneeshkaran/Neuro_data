```python
cap = cv2.VideoCapture('/content/A002_02_BBP_NORMAL_color.avi')

for index, row in df_lm.iterrows():
    frame_no = int(row['Frame'])
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_no)

    ret, frame = cap.read()
    if not ret:
        print(f"Error: Cannot read frame {frame_no}.")
        continue
    for i in range(1, 69):
            x = int(row[f' x{i}'])
            y = int(row[f' y{i}'])
            cv2.circle(frame, (x, y), 1, (255, 255, 255), 1)
            if i > 1 and i < 18 :
                cv2.line(frame, (int(row[f' x{i-1}']), int(row[f' y{i-1}'])), (x, y), (255, 255, 255),1)
            elif i > 18 and i < 23 :
                cv2.line(frame, (int(row[f' x{i-1}']), int(row[f' y{i-1}'])), (x, y), (255, 255, 255),1)
            elif i > 23 and i < 28 :
                cv2.line(frame, (int(row[f' x{i-1}']), int(row[f' y{i-1}'])), (x, y), (255, 255, 255),1)
            elif i > 28 and i < 32 :
                cv2.line(frame, (int(row[f' x{i-1}']), int(row[f' y{i-1}'])), (x, y), (255, 255, 255),1)
            elif i > 32 and i < 37 :
                cv2.line(frame, (int(row[f' x{i-1}']), int(row[f' y{i-1}'])), (x, y), (255, 255, 255),1)
            elif i > 37 and i < 43 :
                cv2.line(frame, (int(row[f' x{i-1}']), int(row[f' y{i-1}'])), (x, y), (255, 255, 255),1)
            elif i > 43 and i < 49 :
                cv2.line(frame, (int(row[f' x{i-1}']), int(row[f' y{i-1}'])), (x, y), (255, 255, 255),1)
            elif i > 49 and i < 61 :
                cv2.line(frame, (int(row[f' x{i-1}']), int(row[f' y{i-1}'])), (x, y), (255, 255, 255),1)
            elif i > 61 and i <= 68 :
                cv2.line(frame, (int(row[f' x{i-1}']), int(row[f' y{i-1}'])), (x, y), (255, 255, 255),1)
    print(f"Displaying frame {frame_no}")
    cv2_imshow(frame)
    cv2.waitKey(20)

cap.release()
```

```python
cap=cv2.VideoCapture('/content/A002_02_BBP_NORMAL_color.avi')
for index, row in df_bb.iterrows():
    frame_no = int(row['Frame'])
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_no)

    ret, frame = cap.read()
    if not ret:
        print(f"Error: Cannot read frame {frame_no}.")
        continue
    x1=int(row[' x1'])
    y1=int(row[' y1'])
    x2=int(row[' x2'])
    y2=int(row[' y2 '])
    cv2.line(frame,(x1,y1),(x1,y2), (255, 255, 255), 1)
    cv2.line(frame,(x2,y2),(x2,y1), (255, 255, 255), 1)
    cv2.line(frame,(x1,y1),(x2,y1), (255, 255, 255), 1)
    cv2.line(frame,(x2,y2),(x1,y2), (255, 255, 255), 1)
    print(f"Displaying frame{frame_no}")
    cv2_imshow(frame)
    cv2.waitKey(20)
cap.release()
```
