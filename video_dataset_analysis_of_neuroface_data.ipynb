```python
import os
import pandas as pd
base_dir = "/content/drive/MyDrive/neuro_data/Neuroface"
categories = ["HC", "PS", "ALS"]

frame_counts = {category: {} for category in categories}

for category in categories:
    landmark_dir = os.path.join(base_dir, category, "Landmarks_gt")

    for file_name in os.listdir(landmark_dir):
        if file_name.endswith(".txt"):
            parts = file_name.split('_')
            if len(parts) >= 4:
                task_name = parts[3].upper()
                file_path = os.path.join(landmark_dir, file_name)
                df = pd.read_csv(file_path)


                frame_count = df.iloc[:, 0].nunique()

                if task_name in frame_counts[category]:
                    frame_counts[category][task_name] += frame_count
                else:
                    frame_counts[category][task_name] = frame_count

for category, counts in frame_counts.items():
    print(f"Frame Counts Per Task for {category}: {counts}")
```

```python
df_ALS_TASK=pd.DataFrame.from_dict(frame_counts['ALS'],orient='index',columns=['ALS'])
df_PS_TASK=pd.DataFrame.from_dict(frame_counts['PS'],orient='index',columns=['PS'])
df_HC_TASK=pd.DataFrame.from_dict(frame_counts['HC'],orient='index',columns=['HC'])
df_ALL_TASK=pd.concat([df_ALS_TASK,df_PS_TASK,df_HC_TASK],axis=1)
df_ALL_TASK
```

```python
df1=df_ALL_TASK.T
df1=df1.reindex(columns=['NORMAL','PA','PATAKA','BIGSMILE','BLOW','BROW','KISS','OPEN','SPREAD'])
df1=df1.rename(columns={'NORMAL':'BBP'})
df1=df1.T
```

```python
plt.bar(df1.index, df1['HC'], label='HC', color='blue')
plt.bar(df1.index, df1['ALS'],
        bottom=df1['HC'], label='ALS', color='red')
plt.bar(df1.index, df1['PS'],
        bottom=df1['HC'] + df1['ALS'], label='PS', color='green')
plt.xlabel('Subtask')
plt.ylabel('No Of frames')
plt.title('NUMBER OF PER TASK DISTIBUTED BY Disease Type ACCORDING TO THE PAPAER')
plt.legend()
plt.xticks(ticks=range(len(df1.index)), labels=df1.index, rotation=45, ha='right')
for i, subtask in enumerate(df1.index):
    plt.text(i, df1['HC'][i] / 2,
             s=df1['HC'][i], ha='center')
    plt.text(i, df1['HC'][i] + df1['ALS'][i] / 2,
             s=df1['ALS'][i], ha='center')
    plt.text(i, df1['HC'][i] + df1['ALS'][i] + df1['PS'][i] / 2,
             s=df1['PS'][i], ha='center')
```

```python
import os
import pandas as pd

df_frame = pd.DataFrame()
base_dir = "/content/drive/MyDrive/neuro_data/Neuroface"
categories = ["HC", "PS", "ALS"]

for category in categories:
    landmark_dir = os.path.join(base_dir, category, "Landmarks_gt")

    for file_name in os.listdir(landmark_dir):
        if file_name.endswith(".txt"):
            parts = file_name.split('_')
            if len(parts) >= 4:
                task_name = parts[3].upper()
                file_path = os.path.join(landmark_dir, file_name)
                df = pd.read_csv(file_path)

                # Add the frame count to df_frame as a new row
                df_frame = pd.concat([df_frame, pd.DataFrame({'file_name': [file_name], 'frame_count': [df.iloc[:, 0].nunique()],'task_name':[task_name]})],ignore_index=True)
print(df_frame)
```
